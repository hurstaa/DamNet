```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns',None)
```

```python
#load inputs

#load site file
sites = pd.read_csv(r"E:\ResSed\MediumResolution_DamLinkages\Manuscript\FinalInputFiles\sites.csv")
#load spatial edits file
spatialedits = pd.read_csv(r"E:\ResSed\MediumResolution_DamLinkages\Manuscript\FinalInputFiles\NID_spatialEdits.csv")


#load damnet to determine which sites belong in the dam attribute file
damnet = pd.read_csv(r"E:\ResSed\MediumResolution_DamLinkages\Manuscript\Outputs_Nov\NIDsMappedInBasins_MedRes_110424.csv")
# damnet = pd.read_csv(r"E:\ResSed\MediumResolution_DamLinkages\Manuscript\Outputs_Melissa\NID_GRanDjoin2.csv")
```

```python
damnet.head()
```

```python
att = pd.merge(sites,spatialedits,how='outer',left_on='NID',right_on='NID_ID',indicator=True)


# att = att[['NID','ShortID','Year_Compl_site','Batch_for','CapOrig_m3','Capm3_p','IsSite','IsUSBR','IsUSACE','IsLock']]
att.loc[att.NID.isna(),'NID'] = att.NID_ID
att.loc[att.Lat.isna(),'Lat'] = att.lat
att.loc[att.Long.isna(),'Long'] = att.long


att.loc[(att['_merge']=='both'),'Lat'] = att.lat #use spatial edits lat and long if a join between both files
att.loc[(att['_merge']=='both'),'Long'] = att.long

att.loc[att.CapOrig_m3.isna(),'CapOrig_m3'] = att.Capm3_p
att.IsSite = att.IsSite.fillna(0)
att.IsRiverMth = att.IsRiverMth.fillna(0)
att.IsUSBR = att.IsUSBR.fillna(0)
att.IsUSACE = att.IsUSACE.fillna(0)
att.delta = att.delta.fillna(0)

att.loc[att['_merge']=='left_only','Moved'] = 1 #assign moved to be 1 for site file
att.Moved = att.Moved.fillna(0)
att.Deleted = att.Deleted.fillna(0)

att = pd.merge(att,damnet[['NID','ShortID']],how = 'left',on='NID')

att['ShortID'] = att['ShortID_x']
att.loc[att.ShortID.isna(),'ShortID'] = att.ShortID_y

att.loc[~att.USBRname.isna(),'Reservoir'] = att.USBRname

att = att.drop(['ShortID_x','ShortID_y','lat','long', 'Unnamed: 0', 'Unnamed: 5','_merge'],axis=1)
# att = att[['NID','ShortID','Reservoir','Lat','Long','Moved','Deleted','Year_Compl_site','YrRemoved','Batch_for','CapOrig_m3','IsSite','IsUSBR','IsUSACE','IsRiverMth','delta','IsLock']]

# att.rename(columns = {'CapOrig_m3':'Capacity_m3','Year_Compl_site':'Year_Compl_user','Batch_for':'DataSource'}, inplace=True)
att.head()

# att.IsSite = 0
# att.loc[(att.IsUSBR == 1)|(att.IsUSACE == 1),'IsSite'] = 1



# att.to_csv(r"E:\ResSed\MediumResolution_DamLinkages\Manuscript\FinalInputFiles\attributes_Mel.csv",index=False)
```
